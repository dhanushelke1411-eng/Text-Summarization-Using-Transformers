```python
from transformers import pipeline

summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)

text = """
Artificial Intelligence (AI) is rapidly transforming industries across the globe.
Machine learning and deep learning technologies are helping businesses automate tasks,
analyze large amounts of data, and improve decision-making processes.
In healthcare, AI assists doctors in disease diagnosis and drug discovery.
In finance, AI is used for fraud detection, stock market analysis, and customer support.
Educational platforms use AI to personalize learning experiences for students.
As AI technology continues to evolve, companies are investing heavily in research and development
to create smarter and more efficient systems that can solve real-world problems.
"""

summary = summarizer(
    text,
    max_length=80,
    min_length=30,
    do_sample=False
)

print("Original Text:\n")
print(text)

print("\nSummarized Text:\n")
print(summary[0]['summary_text'])
```

```python
user_text = input("Enter your long paragraph:\n")

summary = summarizer(
    user_text,
    max_length=80,
    min_length=30,
    do_sample=False
)

print("\nSummarized Text:\n")
print(summary[0]['summary_text'])
```
